

## **Q1. Identify and summarize missing values**

**Solution:**

Using Excel Copilot prompt:

> “Identify columns with missing values and summarize the count of blanks in each column.”

Manual Method:

* Select entire dataset
* Go to **Home → Find & Select → Go To Special → Blanks**
* Or use formula:

```excel
=COUNTBLANK(A:A)
```


## **Q2. Fill missing values in Region and Category columns**

**Solution:**

Using Copilot prompt:

> “Fill missing values in Region and Category using the previous valid entry.”

Manual Method:

1. Select Region column
2. Press **Ctrl + G → Special → Blanks**
3. Type:

```excel
=Above Cell
```

4. Press **Ctrl + Enter**
5. Copy → Paste as Values


## **Q3. Fill missing numeric values (Units Sold & Unit Price)**

**Solution:**

Using Copilot prompt:

> “Fill missing numeric values with the average of the column.”

Manual Formula Example:

If Units Sold is column C:

```excel
=AVERAGE(C:C)
```

Then:

* Select blanks
* Enter:

```excel
=AVERAGE(C:C)
```

* Ctrl + Enter
* Paste as values

Repeat for Unit Price column.

## **Q4. Create new calculated column (Total Revenue)**

**Solution:**

Formula:

```excel
=Units Sold * Unit Price
```

Example:
If Units Sold = C2
If Unit Price = D2

```excel
=C2*D2
```

Drag down.

Name column: **Total Revenue**

## **Q5. Total price of Order ID = ‘ORD0123’**

Using Copilot:

> “Calculate total revenue where Order ID is ORD0123.”

Manual Formula:

```excel
=SUMIF(A:A,"ORD0123",E:E)
```

Where:

* A:A → Order ID column
* E:E → Total Revenue column

## **Q6. Summary chart of average profit by region**

Step 1: Calculate Profit column (if not present)

```excel
=Total Revenue - Cost
```

Step 2:

* Insert → Pivot Table
* Rows → Region
* Values → Profit (Set to Average)

Step 3:

* Insert → Column Chart


## **Q7. Pivot Table: Total Sales by Category and Region**

Steps:

1. Select full dataset
2. Insert → Pivot Table
3. Drag:

   * Rows → Category
   * Columns → Region
   * Values → Total Revenue (Sum)

This gives total sales breakdown.
